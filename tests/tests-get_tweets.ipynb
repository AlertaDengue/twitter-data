```python
def generate_filenames():
    months = [num for num in range(1, 13)]
    years = [num for num in range(2010, 2023)]

    filenames = [];
    for year in years:
        for month in months:
            filenames.append(f'{year}/snscrape_dengue_{year}-{month}.json')
    return filenames
```

```python
def generate_filenames_by_year():
    months = [num for num in range(1, 13)]
    years = [num for num in range(2010, 2023)]
    
    filenames = [];

    for year in years:
        filenames_by_year = [];
        for month in months:
            filenames_by_year.append(f'{year}/snscrape_dengue_{year}-{month}.json')
        filenames.append(filenames_by_year)
    return filenames
```

```python
filenames = generate_filenames_by_year()
```

```python
def return_tweets(filename):
    tweets = []
    for line in open(filename, 'r'):  # replace with "with"
        tweets.append(json.loads(line))
    return tweets
```

```python
def return_tweets_by_year(filenames):
    tweets = []

    for filename in filenames:
        print(f"* tweets in {filename}")
        tweets_by_year = [];
        for line in open(filename, 'r'):  # replace with "with"
            tweets_by_year.append(json.loads(line))
        tweets.append(tweets_by_year)
    return tweets
```

```python
len_tweets_by_year = []

for set_of_filenames in filenames:
    len_tweets = []
    tweets = return_tweets_by_year(set_of_filenames)
    for tweet in tweets:
        len_tweets.append(len(tweet))
    len_tweets_by_year.append(len_tweets)
```

```python
import matplotlib.pyplot as plt

for tweets in len_tweets_by_year:
    plt.plot(tweets, c='k', alpha=0.4)
```
